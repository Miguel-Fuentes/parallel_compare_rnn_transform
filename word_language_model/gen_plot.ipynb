```python
data = pd.DataFrame({
    'lstm4' : lstm4,
    'lstm8' : lstm8,
    'lstm16' : lstm16,
    'lstm_gpu' : lstm_gpu,
    'transform4' : transform4,
    'transform8' : transform8,
    'transform16' : transform16,
    'transform_gpu' : transform_gpu
})
```

```python
total_time = pd.DataFrame({
    'LSTM 4 Thread' : sum(lstm4),
    'LSTM 8 Thread' : sum(lstm8),
    'LSTM 16 Thread' : sum(lstm16),
    'LSTM GPU' : sum(lstm_gpu),
    'Transformer 4 Thread' : sum(transform4),
    'Transformer 8 Thread' : sum(transform8),
    'Transformer 16 Thread' : sum(transform16),
    'Transformer GPU' : sum(transform_gpu),
}, index={0})
```

```python
total_time
```

```python
sns.set(style="whitegrid", font_scale=3)

plt.figure(figsize=(19,25))
g = sns.barplot(data=total_time)
g.set_xticklabels(g.get_xticklabels(), rotation=25, ha="right")
g.set(xlabel='Model', ylabel='Training Time (seconds)')
fig = g.get_figure()
fig.savefig('results.png')
```
